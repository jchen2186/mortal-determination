```python
def read_clean_convert(path):
    df = pd.read_excel(path)
    df.columns = df.iloc[2]
    df = df.drop(list(range(5)))

    #df['Average Adjusted gross income (AGI)'] = df['Adjusted gross income (AGI) [2]']/df['Number of returns']
    
    columns = list(df.columns)
    for i in range(len(columns)):
        columns[i] = str(columns[i])
        if '[' in columns[i]:
            columns[i] = columns[i][:-4]
        if '\n' in columns[i]:
            columns[i] = ' '.join([substring.strip() for substring in columns[i].split('\n')])
        if columns[i] == 'nan':
            base = columns[i-1]
            columns[i-1] = base + ' Number of returns'
            columns[i] = base + ' Amount'
    df.columns = columns
    
    new_path = path[4:-3] + 'csv'
    df.to_csv(new_path, index=False)
```

```python
for yearly_data_dir in os.listdir(data_dir):
    try:
        os.mkdir((os.path.join(data_dir,yearly_data_dir))[4:])
        for file in os.listdir(yearly_data_dir):
            path = os.path.join(yearly_data_dir, file)
            if file[-4:] == '.xls':
                read_clean_convert(path)
            elif file[-4:] == '.csv':
                try:
                    df = pd.read_csv(path)
                    df.to_csv(path[4:], index=False)
                except UnicodeDecodeError:
                    pass
    except OSError:
        pass
```
