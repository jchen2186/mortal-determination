# Data_load for our visulaization/ Data cleaning


```python
data_load= data.loc[:33,['Year','Live Births','Rate','Fertility Rate']]
data_load
```

```python
data_load['Year'] = data_load.Year.astype( object)
```

```python
data_load.dtypes
```

```python
data_load['Rate'] = data_load.Rate.astype(float)
```

```python
data_load.dtypes
```

```python
data_load.describe()
```

```python
data_load.plot(x='Year',y='Rate',kind='line',color='purple')
plt.ylabel('Birth Rate')
plt.show()
```

```python
data_load.plot(x='Year',y='Live Births',kind='line',color='purple')
plt.ylabel('Live Biths')
plt.show()
```

Special Great recession effect on birth

```python
ss1=data_load.tail(9)
ss1
```

```python
ss1.plot(x='Year',y='Rate',kind='line',color='purple')
plt.ylabel('Birth Rate')
plt.show()
```
